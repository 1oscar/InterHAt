```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,4,8, 12,16]
criteo_auc = [0.782, 0.7828, 0.7835, 0.7845, 0.7825, 0.7825]
criteo_lls = [0.4608, 0.4596, 0.4583, 0.4578, 0.4643, 0.4594]

avazu_auc = [0.7577, 0.758, 0.7585, 0.7582, 0.7578, 0.7588]
avazu_lls = [0.3919, 0.3913, 0.391, 0.391, 0.3914, 0.3908]

frappe_auc = [0.9711, 0.9692, 0.9648, 0.9611, 0.9621, 0.9672]
frappe_lls = [0.1916, 0.2006, 0.2075, 0.2226, 0.2191, 0.2023]
```

```python
font_size=40
fig, ((ax1, ax2), (ax3, ax4), (ax5, ax6)) = plt.subplots(3,2, sharex=True, sharey=False, figsize=(36, 16))
plt.rcParams.update({'font.size': font_size})

end = 5

ax1.plot(x[:end], criteo_auc[:end], linewidth=4, color='#008000')
ax1.set_ylim(0.7810, 0.7850)
ax1.set_yticks([0.7810,0.7820, 0.7830, 0.7840,0.7850])
ax1.set_xlabel("AUC on Criteo")
ax1.grid(True)

ax2.plot(x[:end], criteo_lls[:end], linewidth=4, color="#1e90ff")
ax2.set_ylim(0.455, 0.465)
ax2.set_xlabel("Logloss on Criteo")
ax2.grid(True)

ax3.plot(x[:end], avazu_auc[:end], linewidth=4, color='#008000')
ax3.set_ylim(0.7575, 0.7590)
ax3.set_yticks([0.7575, 0.7580, 0.7585, 0.7590])
ax3.set_xlabel("AUC on Avazu")
ax3.grid(True)

ax4.plot(x[:end], avazu_lls[:end], linewidth=4, color='#1e90ff')
ax4.set_xlabel("Logloss on Avazu")
ax4.set_ylim(0.3905, 0.3920)
ax4.set_yticks([0.3905, 0.3910, 0.3915, 0.3920])
ax4.grid(True)

ax5.plot(x[:end], frappe_auc[:end], linewidth=4, color='#008000')
ax5.set_xlabel("AUC on Frappe")
ax5.set_ylim(0.96, 0.975)
ax5.set_yticks([0.96, 0.965, 0.97, 0.975])
ax5.grid(True)

ax6.plot(x[:end], frappe_lls[:end], linewidth=4, color='#1e90ff')
ax6.set_ylim(0.190, 0.23)
ax6.set_yticks([0.19, 0.2, 0.21, 0.22, 0.23])
ax6.set_xlabel("Logloss on Frappe")
ax6.grid(True)
```
